```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Define paths
annotations_path = "/content/drive/MyDrive/processed_data/processed_data/annotations"
images_path = "/content/drive/MyDrive/processed_data/processed_data/images"


# Create train and val directories
train_images_path = os.path.join(images_path, 'train')
val_images_path = os.path.join(images_path, 'val')
train_annotations_path = os.path.join(annotations_path, 'train')
val_annotations_path = os.path.join(annotations_path, 'val')

os.makedirs(train_images_path, exist_ok=True)
os.makedirs(val_images_path, exist_ok=True)
os.makedirs(train_annotations_path, exist_ok=True)
os.makedirs(val_annotations_path, exist_ok=True)

# Collect all image and annotation files
image_files = []
for root, _, files in os.walk(images_path):
    for file in files:
        if file.endswith(".jpg"):
            image_files.append(os.path.join(root, file))

# Split into train and val
train_files, val_files = train_test_split(image_files, test_size=0.2, random_state=42)

# Function to create the directory structure if it doesn't exist
def ensure_dir(file_path):
    directory = os.path.dirname(file_path)
    if not os.path.exists(directory):
        os.makedirs(directory)

# Copy files to respective directories
for file in train_files:
    relative_path = os.path.relpath(file, images_path)
    destination_path = os.path.join(train_images_path, relative_path)
    ensure_dir(destination_path)
    shutil.copy(file, destination_path)
    annotation_file = os.path.join(annotations_path, relative_path.replace(".jpg", ".txt"))
    if os.path.exists(annotation_file):
        annotation_destination_path = os.path.join(train_annotations_path, relative_path.replace(".jpg", ".txt"))
        ensure_dir(annotation_destination_path)
        shutil.copy(annotation_file, annotation_destination_path)

for file in val_files:
    relative_path = os.path.relpath(file, images_path)
    destination_path = os.path.join(val_images_path, relative_path)
    ensure_dir(destination_path)
    shutil.copy(file, destination_path)
    annotation_file = os.path.join(annotations_path, relative_path.replace(".jpg", ".txt"))
    if os.path.exists(annotation_file):
        annotation_destination_path = os.path.join(val_annotations_path, relative_path.replace(".jpg", ".txt"))
        ensure_dir(annotation_destination_path)
        shutil.copy(annotation_file, annotation_destination_path)

print("Dataset split into train and val.")
```

```python
import os
import shutil
from glob import glob

def flatten_directory(src_dir, dst_dir):
    os.makedirs(dst_dir, exist_ok=True)
    for dirpath, _, filenames in os.walk(src_dir):
        for file in filenames:
            if file.endswith('.txt'):
                shutil.copy(os.path.join(dirpath, file), os.path.join(dst_dir, file))
            else:
                shutil.copy(os.path.join(dirpath, file), os.path.join(dst_dir, file))

base_path = '/content/drive/MyDrive/processed_data/processed_data/'
annotations_train = os.path.join(base_path, 'annotations/training')
annotations_val = os.path.join(base_path, 'annotations/val')
images_train = os.path.join(base_path, 'images/training')
images_val = os.path.join(base_path, 'images/val')

flatten_directory(annotations_train, os.path.join(base_path, 'labels/train'))
flatten_directory(annotations_val, os.path.join(base_path, 'labels/val'))
flatten_directory(images_train, os.path.join(base_path, 'images/train'))
flatten_directory(images_val, os.path.join(base_path, 'images/val'))
```

```python
import os

# Define the annotations path
annotations_path = "/content/drive/MyDrive/processed_data/processed_data/annotations"

# Function to extract class names from annotation files
def get_class_names(annotations_path):
    class_names = set()
    for root, _, files in os.walk(annotations_path):
        for file in files:
            if file.endswith(".txt"):
                file_path = os.path.join(root, file)
                try:
                    with open(file_path, 'r') as f:
                        for line in f:
                            if len(line.strip()) == 0:
                                continue
                            class_name = line.split()[0]
                            class_names.add(class_name)
                    # Debugging information
                    print(f"Processed file: {file_path}")
                except Exception as e:
                    print(f"Error processing file {file_path}: {e}")
    return list(class_names)

# Get class names
class_names = get_class_names(annotations_path)
print("Class names:", class_names)
print("Number of classes:", len(class_names))
```

```python
import os

# Define the paths to your dataset images and annotations directories
dataset_images_dir = '/content/drive/MyDrive/processed_data/processed_data/images/train'
annotations_dir = '/content/drive/MyDrive/processed_data/processed_data/annotations/train'

# Iterate through each subdirectory (assumed to be class folders)
for class_folder in os.listdir(dataset_images_dir):
    class_folder_path = os.path.join(dataset_images_dir, class_folder)
    annotations_class_folder_path = os.path.join(annotations_dir, class_folder)

    if os.path.isdir(class_folder_path) and os.path.isdir(annotations_class_folder_path):
        print(f"Checking label files for images in {class_folder_path}...")

        # Iterate through each image directory
        for image_folder in os.listdir(class_folder_path):
            image_folder_path = os.path.join(class_folder_path, image_folder)
            annotations_image_folder_path = os.path.join(annotations_class_folder_path, image_folder)

            if os.path.isdir(image_folder_path) and os.path.isdir(annotations_image_folder_path):
                # Check for label files (.txt) corresponding to each image
                missing_labels = []
                for image_file in os.listdir(image_folder_path):
                    if image_file.endswith('.jpg') or image_file.endswith('.png'):
                        label_file = os.path.splitext(image_file)[0] + '.txt'
                        label_file_path = os.path.join(annotations_image_folder_path, label_file)
                        if not os.path.exists(label_file_path):
                            missing_labels.append(image_file)
                if missing_labels:
                    print(f"  Missing label files for images in {image_folder_path}:")
                    for missing_image in missing_labels:
                        print(f"    {missing_image}")
                else:
                    print(f"  All images in {image_folder_path} have corresponding label files.")
            else:
                print(f"  Skipping image directory: {image_folder_path} or annotations directory: {annotations_image_folder_path}")
    else:
        print(f"Skipping class folder without annotations: {class_folder_path} or {annotations_class_folder_path}")
```

```python
import os
import shutil

# Define the paths to your dataset images and annotations directories
dataset_images_dir_train = '/content/drive/MyDrive/processed_data/processed_data/images/train'
dataset_images_dir_val = '/content/drive/MyDrive/processed_data/processed_data/images/val'
annotations_dir_train = '/content/drive/MyDrive/processed_data/processed_data/annotations/train'
annotations_dir_val = '/content/drive/MyDrive/processed_data/processed_data/annotations/val'

# Function to check missing labels and delete images
def delete_images_without_labels(images_dir, annotations_dir):
    for class_folder in os.listdir(images_dir):
        class_folder_path = os.path.join(images_dir, class_folder)
        if os.path.isdir(class_folder_path):
            print(f"Checking label files for images in {class_folder_path}...")
            # Iterate through each image directory
            for image_folder in os.listdir(class_folder_path):
                image_folder_path = os.path.join(class_folder_path, image_folder)
                if os.path.isdir(image_folder_path):
                    # Check for label files (.txt) corresponding to each image
                    missing_labels = []
                    for image_file in os.listdir(image_folder_path):
                        if image_file.endswith('.jpg') or image_file.endswith('.png'):
                            label_file = os.path.splitext(image_file)[0] + '.txt'
                            label_file_path = os.path.join(annotations_dir, class_folder, image_folder, label_file)
                            if not os.path.exists(label_file_path):
                                missing_labels.append(image_file)

                    # If missing labels, delete images
                    if missing_labels:
                        print(f"  Deleting images without labels in {image_folder_path}:")
                        for missing_image in missing_labels:
                            image_path = os.path.join(image_folder_path, missing_image)
                            os.remove(image_path)
                            print(f"    Deleted: {missing_image}")
                    else:
                        print(f"  All images in {image_folder_path} have corresponding label files.")
                else:
                    print(f"  Skipping non-directory: {image_folder_path}")
        else:
            print(f"Skipping non-directory: {class_folder_path}")

# Delete images without labels in train and val datasets
delete_images_without_labels(dataset_images_dir_train, annotations_dir_train)
delete_images_without_labels(dataset_images_dir_val, annotations_dir_val)
```
